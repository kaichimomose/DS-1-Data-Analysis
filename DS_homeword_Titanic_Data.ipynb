```python
path = "datasets/titanic.csv"
df = pd.read_csv(path)
df.describe()
```

```python
df
```

```python
"1. Given that a passenger is under 30 but over 20, what are the chances that they are in first class?"
```

```python
U30_O20_df = df[(df['Age'] > 20.0) & (df['Age'] <30.0)]
U30_O20_df.describe()
```

```python
firstclass_U30_O20_df = U30_O20_df[U30_O20_df['Pclass'] == 1]
firstclass_U30_O20_df.describe()
```

```python
fig = plt.figure()

for frame in [U30_O20_df, firstclass_U30_O20_df]:
    plt.hist(frame['Age'], bins=5)

plt.xlim(20,30)
plt.ylim(0,60)
plt.show()
```

```python
firstclass_U30_O20_chance = len(firstclass_U30_O20_df)/len(U30_O20_df)
firstclass_U30_O20_chance
```

```python
"2. Given that  a passenger is in 1st class, what are the chances that they paid less than 50$ for their ticket"
```

```python
first_class_df = df[df['Pclass'] == 1]
first_class_df.describe()
```

```python
less_50_first_class_df = first_class_df[first_class_df['Fare'] < 50]
less_50_first_class_df.describe()
```

```python
fig = plt.figure()

for frame in [first_class_df, less_50_first_class_df]:
    plt.bar(frame['Pclass'])

plt.xlim(0,2)
plt.ylim(0,250)
plt.show()
```

```python
less_50_first_class_chance = len(less_50_first_class_df)/len(first_class_df)
less_50_first_class_chance
```

```python
"4. GIVEN that a survivor is from Queenstown, did s/he have a family?"
```

```python
survivor_Queenstown_df = df[(df['Survived'] == 1) & (df['Embarked'] == 'Q')]
survivor_Queenstown_df.describe()
```

```python
with_family_survivor_Queenstown_df = survivor_Queenstown_df[(survivor_Queenstown_df['SibSp'] > 0) | (survivor_Queenstown_df['Parch'] > 0)]
with_family_survivor_Queenstown_df.describe()
```

```python
with_family_survivor_Queenstown_chance = len(with_family_survivor_Queenstown_df)/len(survivor_Queenstown_df)
with_family_survivor_Queenstown_chance
```
