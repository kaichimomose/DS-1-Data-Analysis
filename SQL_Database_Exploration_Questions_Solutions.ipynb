# 1. Get the player names, player hometowns, school names, player positions, and school conferences for all players with position ‘QB’ or ‘WR’
```SQL
select player_name, hometown, players.school_name, position, conference from benn.college_football_players players
join benn.college_football_teams teams on players.school_name = teams.school_name
where position = 'QB' or position = 'WR'
```

# 2. Create a table that displays the average height of players at each position
```SQL
select avg(height) as avg_height, position from benn.college_football_players
where height != 0
group by position
order by avg_height desc
```

# 3. Write a query that find the top 5 tallest players, as well as their position, height, school, and the conference they belong to. 
```SQL
select player_name, position, height, players.school_name, conference from benn.college_football_players players
join benn.college_football_teams teams on players.school_name=teams.school_name
order by height desc
limit 5
```

# 5. Companies from which 10 cities have the highest average acquisition amount?
```SQL
select city, avg(price_amount) average from tutorial.crunchbase_companies companies
join tutorial.crunchbase_acquisitions acquisitions on companies.name=acquisitions.company_name
where price_amount > 0
group by city
order by average desc
limit 10
```
