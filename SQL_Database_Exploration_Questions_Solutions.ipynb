````python
dataset = df["funding_total_usd"]
dataset
````

````python
def get_sample(dataset, n=30):
    """Grabs a random subsample of size 'n' from dataset.
    Outputs the mean of the subsample."""
    sample = np.random.choice(dataset, size=n)
    return np.mean(sample)

def create_sample_distribution(dataset, size=100000):
    """Creates a dataset of subsample means.  The length of the dataset is specified by the 'size' 
    keyword argument. Should return the entire sample distribution as a numpy array.  """
    sample_averages = np.array([])
    for i in range(0, size):
        average = get_sample(dataset)
        sample_averages = np.append(sample_averages, average)   
    return sample_averages
````

````python
sample_averages = create_sample_distribution(dataset)
plt.hist(sample_averages)
````

````python
mean = np.mean(dataset)
mean
````

````python
std = np.std(sample_averages)
std
````

````python
import scipy.stats as st
````

````python
left_tail_of_25m = st.norm.cdf((25000000-mean)/std)
left_tail_of_30m = st.norm.cdf((30000000-mean)/std)
print("Area to the left of 25million: {}".format(left_tail_of_25m))
print("Area to the left of 30million: {}".format(left_tail_of_30m))
print("Answer : {}%".format((left_tail_of_30m - left_tail_of_25m)*100))
````
